```python

import pandas as pd
```

```python

parent= pd.read_excel (r'Parent Table.xlsx')
```

```python
parent
```

```python
imp = pd.read_excel(r'Import Discount Table.xlsx')
exp = pd.read_excel(r'Expot tax Table.xlsx')
```

```python
imp
```

```python
exp
```

```python
output = pd.DataFrame(columns = ['VegetableName', 'ImportAmount','ExportAmount'])
```

```python
output
```

```python
combined=pd.concat([parent,imp,exp],axis=1)
```

```python
combined
```

```python
combined=combined.T.drop_duplicates().T
```

```python
combined
```

```python
for i in combined.index:
  output.loc[i,'VegetableName']=combined.loc[i,'Vegetable']
  output.loc[i,'ImportAmount']=(int(combined.loc[i,'Import Rate'])-int(combined.loc[i,'Discount Amount (Rs)']))
  output.loc[i,'ExportAmount']=(combined.loc[i,'Export Rate']+(combined.loc[i,'Tax Percentage']*combined.loc[i,'Export Rate']))
```

```python
output
```

```python
output.to_json("output.json", orient = 'records')
```
